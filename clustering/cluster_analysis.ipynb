```python
import pandas as pd
import plotly.express as px
from sklearn.manifold import TSNE
```

```python
playlist_embeddings = pd.read_parquet("playlist_embeddings.parquet")
playlist_meta = pd.read_parquet("../data/parquet/playlist_metadata.parquet")
playlist_embeddings.head()
```

```python
# cluster visualization
tsne_components = TSNE(n_components=3, perplexity=10, random_state=42).fit_transform(playlist_embeddings[[f"emb_{i}" for i in range(0, 32)]])
playlist_embeddings["tsne1"] = tsne_components[:, 0]
playlist_embeddings["tsne2"] = tsne_components[:, 1]
playlist_embeddings["tsne3"] = tsne_components[:, 2]
playlist_embeddings["cluster"] = playlist_embeddings["cluster"].astype(str)

fig = px.scatter_3d(
    playlist_embeddings,
    x='tsne1',
    y='tsne2',
    z='tsne3',
    color='cluster',
    title='Clusters Visualized',
    color_discrete_sequence=px.colors.qualitative.Bold
)

fig.write_html("PlaylistClusters.html")
```

```python
playlist_meta["pid"] = playlist_meta["pid"].astype(str)
data = playlist_meta.merge(playlist_embeddings[["node", "cluster"]], left_on="pid", right_on="node", how="inner")
```

```python
for i in range(0, 19):
    print(f"Cluster {i} Playlist Names: {data[data['cluster'] == i]['name'].tolist()}" )
```
